```python
import numpy as np
# a = np.arange(9).reshape(3,3) + 10
a = np.array([[10, 20], [5, 25], [15, 8]])
print(a)
print()
print(np.argmin(a))
print(np.argmin(a, axis=0))
print(np.argmin(a, axis=1))
print()
print(a.shape[0])
print(a.mean(axis=0))
print(np.median(a, axis=0))
```

# [0 0 1] - Radvis, 3 rader (10,20), (5,25) och (15,8), lägsta värdet i 1:a rad '10' kol 0, 2:a rad '5' kol 0 och 3:e rad '8' kol 1
